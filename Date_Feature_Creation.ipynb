```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Superstore.csv")
```

```python
print(df.isnull().sum())
```

```python
print(df.head())
```

```python
df.drop("Postal Code",axis=1,inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.drop(["Row ID","Order ID","Customer ID","Customer Name","Product ID","Product Name"],axis=1,inplace=True)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],format="%d/%m/%Y")
df["Ship Date"]=pd.to_datetime(df["Ship Date"],format="%d/%m/%Y")
```

```python
print(df.head())
```

Feature extraction

```python
df["Order_Year"]=df["Order Date"].dt.year
df["Order_Month"]=df["Order Date"].dt.month
df["Order_Day"]=df["Order Date"].dt.day
df["Order_Day_Of_Week"]=df["Order Date"].dt.day_of_week
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
df["Ship_year"] = df["Ship Date"].dt.year
df["Ship_Month"]=df["Ship Date"].dt.month
df["Ship_Day"]=df["Ship Date"].dt.day
df["Ship_Day_Of_Week"]=df["Ship Date"].dt.day_of_week
```

```python
print(df.head())
```

```python
df["Shipping_Days"] = df["Ship_Day"] - df["Order_Day"]
```

```python
print(df.head())
```

```python
df.to_csv("featureExtration_SuperStore.csv",index=False)
```
